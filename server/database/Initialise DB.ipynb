For dropping tables and reset. 

```python
from psycopg2 import sql
table1 = "NUH_Data"
table2 = "SGH_Data"
table3 = "SingHealth_Data"
sqlQuery = sql.SQL("""
DROP TABLE {table1}, {table2}, {table3};
""").format(
    table1=sql.Identifier(table1),
    table2=sql.Identifier(table2),
    table3=sql.Identifier(table3)
)
rdsDB.masterCursor.execute(sqlQuery)
```

```python
org1 = "NUH"
org2 = "SGH"
org3 = "SingHealth"
sqlQuery = sql.SQL("""
REVOKE ALL PRIVILEGES ON ALL TABLES IN SCHEMA public FROM {role1}, {role2}, {role3};
REVOKE ALL PRIVILEGES ON ALL SEQUENCES IN SCHEMA public FROM {role1}, {role2}, {role3};
REVOKE ALL ON DATABASE postgres FROM {role1}, {role2}, {role3};
DROP ROLE {role1}, {role2}, {role3};
""").format(
    role1=sql.Identifier(org1),
    role2=sql.Identifier(org2),
    role3=sql.Identifier(org3)
)
rdsDB.masterCursor.execute(sqlQuery)
```

```python
sqlQuery = sql.SQL("""
DROP USER {user};
""").format(user=sql.Identifier('Fred'))
rdsDB.masterCursor.execute(sqlQuery)
```
